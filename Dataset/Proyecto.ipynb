```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('C:\\Users\\sony\\Desktop\\MAESTRÍA INTELIGENCIA ARTIFICIAL\\3_Aprendizaje Automatico\\Introduccion_Aprendizaje_Automatico_2025\\Dataset\\apartments.csv', 
                 on_bad_lines='skip', encoding='latin1')

# Verificar si la columna 'cityname' existe
if 'cityname' in df.columns:
    conteo_ciudades = df['cityname'].value_counts()

    print("Ciudades más frecuentes:")
    print(conteo_ciudades.head(10))  # Mostrar las 10 primeras

    # Graficar
    plt.figure(figsize=(10, 6))
    conteo_ciudades.head(10).plot(kind='bar', color='skyblue')
    plt.title("Ciudades más frecuentes")
    plt.xlabel("Ciudad")
    plt.ylabel("Cantidad")
    plt.show()

else:
    print("La columna 'cityname' no se encuentra en el archivo. Las columnas disponibles son:")
    print(df.columns.tolist())  # Mostrar las columnas disponibles
   
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\sony\\Desktop\\MAESTRÍA INTELIGENCIA ARTIFICIAL\\3_Aprendizaje Automatico\\Introduccion_Aprendizaje_Automatico_2025\\Dataset\\apartments.csv', sep=';', on_bad_lines='skip', encoding='latin1')

if 'state' in df.columns:
   
    conteo_estados = df['state'].value_counts()

    print("Porcentaje de departamentos en renta por estado:")
    print(conteo_estados.head(15))

    plt.figure(figsize=(8, 8))
    conteo_estados.head(15).plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors,
        wedgeprops={'edgecolor': 'white'}
    )
    plt.title("Porcentaje de departamentos en renta por estado")
    plt.ylabel("")  
    plt.tight_layout()
    plt.show()

else:
    print("La columna 'state' no se encuentra en el archivo. Las columnas disponibles son:")
    print(df.columns.tolist())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('C:\\Users\\sony\\Desktop\\MAESTRÍA INTELIGENCIA ARTIFICIAL\\3_Aprendizaje Automatico\\Introduccion_Aprendizaje_Automatico_2025\\Dataset\\apartments.csv', sep=';', on_bad_lines='skip', encoding='latin1')


print("Columnas disponibles en el DataFrame:")
print(df.columns)


plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')  
plt.title("Distribución del Precio")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(df['bedrooms'], bins=30, kde=True, color='green')
plt.title("Distribución de la Cantidad de Habitaciones")
plt.xlabel("Número de Habitaciones")
plt.ylabel("Frecuencia")
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bedrooms', y='price', color='red')
plt.title("Relación entre Precio y Número de Habitaciones")
plt.xlabel("Número de Habitaciones")
plt.ylabel("Precio")
plt.show()


plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='state', y='price', hue='state', palette='Set2', showfliers=False)
plt.title("Distribución de Precios por Estado")
plt.xlabel("Estado")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.legend(title="Estado", bbox_to_anchor=(1.05, 1), loc='upper left')  # Mostrar leyenda fuera del gráfico
plt.show()
```
