```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Dataset/apartments.csv', sep=';', on_bad_lines='skip', encoding='latin1')

if 'cityname' in df.columns: #exite la columna?
    conteo_ciudades = df['cityname'].value_counts()

    print("Ciudades más frecuentes:") #primero mostramos la tabla
    print(conteo_ciudades.head(10))

    plt.figure(figsize=(10, 6))#ahora grafico
    conteo_ciudades.head(10).plot(kind='bar', color='skyblue')
    plt.title("Ciudades más frecuentes")
    plt.xlabel("Ciudad")
    plt.ylabel("Cantidad")
    plt.show()
    
else:
    print("La columna 'cityname' no se encuentra en el archivo. Las columnas disponibles son:")
    print(df.columns.tolist())
   
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset/apartments.csv', sep=';', on_bad_lines='skip', encoding='latin1')

if 'state' in df.columns:
   
    conteo_estados = df['state'].value_counts()

    print("Porcentaje de departamentos en renta por estado:")
    print(conteo_estados.head(15))

    plt.figure(figsize=(8, 8))
    conteo_estados.head(15).plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors,
        wedgeprops={'edgecolor': 'white'}
    )
    plt.title("Porcentaje de departamentos en renta por estado")
    plt.ylabel("")  
    plt.tight_layout()
    plt.show()

else:
    print("La columna 'state' no se encuentra en el archivo. Las columnas disponibles son:")
    print(df.columns.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("apartments.csv") 


plt.figure(figsize=(8,5))
plt.hist(df["price"], bins=50, edgecolor="black", color="orange")
plt.title("Distribución de precios de renta (todos los datos)")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()


df_filtered = df[df["price"] < 5000]  


plt.figure(figsize=(8,5))
plt.hist(df_filtered["price"], bins=40, edgecolor="black", color="skyblue")
plt.title("Distribución de precios de renta (sin outliers)")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()
```
